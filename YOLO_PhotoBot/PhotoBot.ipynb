```
import telebot
import urllib.request
from imageai.Detection import ObjectDetection
import os 

token = ''
execution_path = os.getcwd()
detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("yolo.h5")
detector.loadModel()

bot = telebot.TeleBot(token)

@bot.message_handler(commands = ['start'])
def welcome_user(message):
  bot.reply_to(message, "Привіт, завантаж картинку")


@bot.message_handler(content_types=['photo'])
def serach_object(message):
  file_info = bot.get_file(message.photo[len(message.photo)-1].file_id)
  downloaded_file = bot.download_file(file_info.file_path)
  scr = 'files/'+file_info.file_path
  with open (scr, 'wb') as new_file:
    new_file.write(downloaded_file)
    bot.reply_to(message, "Завантаженно, очікуйте")
    detections = detector.detectObjectsFromImage(input_image=scr,
                                                 output_image_path = "final.jpg",
                                                 minimum_percentage_probability = 30)
    bot.reply_to(message, "На картинці знайденно:")
    for eachObject in detections:
      result = str(eachObject["name"]) + ":" + str(eachObject["percentage_probability"])
      bot.reply_to(message, result)

    photo = open('final.jpg','rb')
    bot.send_photo(message.chat.id, photo)

bot.polling()
```
